```python
def get_car_type_id(row):
    car_type = row['car_type']
    if car_type == 'Pick Up':
        return 3
    car_type = car_type.replace(' ', '')
    if car_type == 'SUV/Crossover':
        return 0
    elif car_type == 'Sedan':
        return 1
    elif car_type == 'Coupe':
        return 2
    elif car_type == 'Convertible':
        return 4
    elif car_type == 'Hatchback':
        return 5
    elif car_type == 'Van':
        return 6
    else:
        return 7
```

```python
mileage_distribution = [0, 10000, 50000, 100000, 190000]
mileage_distribution_id = [0, 1, 2, 3, 4]
mileage_distribution = df['mileage'].value_counts()

def get_mileage_distribution_id(row):
    if row['mileage'] < 50.000:
        return 1
    else:
        return 0
```

```python
premium_manufacturers = ['Audi', 'BMW', 'Mercedes-Benz', 'Porsche', 'Land Rover', 'Lexus', 'Infiniti', 'Cadillac', 'Volvo', 'Alfa Romeo', 'Jaguar', 'Rolls-Royce', 'Bentley', 'Ferrari', 'Lamborghini', 'Maserati', 'Aston Martin']
```

```python
def get_premium_manufacturer_id(row):
    car = row['car_manufacturer']
    if car in premium_manufacturers:
        return 1
    else:
        return 0
```

```python
def get_fuel_type_id(row):
    fuel_type = row['fuel_type']
    if fuel_type == 'Gasoline':
        return 1
    elif fuel_type == 'Diesel':
        return 1
    elif fuel_type == 'Hybrid':
        return 2
    elif fuel_type == 'Electric':
        return 3
    else:
        return 4
```

```python
import datetime

def get_time_owned(row):
    release = row['model_year']
    now = datetime.datetime.now().year
    owned =  now - release
    if owned <= 1:
        return 1
    else:
        return 0
```

```python
from sklearn.preprocessing import OneHotEncoder
import sklearn as sk

ohe = OneHotEncoder(handle_unknown='ignore', sparse_output=False).set_output(transform='pandas')
```

```python
fuel_type_ohe = ohe.fit_transform(df[['fuel_type']])
fuel_type_ohe
```

```python
car_type_ohe = ohe.fit_transform(df[['car_type']])
car_type_ohe
```

```python
location_ohe = ohe.fit_transform(df[['car_location']])
location_ohe
```

```python
def get_per_year_cost(row):
    release = row['model_year']
    now = datetime.datetime.now().year
    time_owned =  now - release
    if time_owned == 0:
        return row['price']
    else:
        return row['price'] / time_owned
```

```python
def get_per_kkm_cost(row):
    mileage = row['mileage']
    if mileage == 0:
        return row['price']
    else:
        return row['price'] / mileage
```

```python
def car_type_suv(row):
    car_type = row['car_type']
    car_type = car_type.replace(' ', '')
    if car_type == 'SUV/Crossover':
        return 1
    else:
        return 0
```

```python
def car_interior_rare(row):
    car_interior = row['interior_color']
    if car_interior == 'Black' or car_interior == 'Beige':
        return 0
    else:
        return 1
```

```python
def bins_price(row):
    bins = [10000, 30000, 50000, 70000, 100000, 150000, 200000, 300000, 500000, 1000000, 100000000]
    return np.digitize(row['price'], bins)
```

### Data was preprocessed, used one hot encoding, binning, and some new features with strong correlation with price were added:

```python
preprocessed_df = pd.DataFrame()
preprocessed_df['price_bins'] = df.apply(bins_price, axis=1)
preprocessed_df['model_year'] = df['model_year']
preprocessed_df['cylinders'] = df['cylinders']
preprocessed_df['wheels_size'] = df['wheels_size']
preprocessed_df['time_owned'] = df.apply(get_time_owned, axis=1)
preprocessed_df['per_year_cost'] = df.apply(get_per_year_cost, axis=1).astype(int)
preprocessed_df['per_kkm_cost'] = df.apply(get_per_kkm_cost, axis=1).astype(int)
preprocessed_df['car_type_suv'] = df.apply(car_type_suv, axis=1)
preprocessed_df['car_interior_rare'] = df.apply(car_interior_rare, axis=1)
preprocessed_df['fuel_type_id'] = df.apply(get_fuel_type_id, axis=1)
preprocessed_df['car_type_id'] = df.apply(get_car_type_id, axis=1)
preprocessed_df['mileage_distribution_id'] = df.apply(get_mileage_distribution_id, axis=1)
preprocessed_df['premium_manufacturers'] = df.apply(get_premium_manufacturer_id, axis=1)
preprocessed_df = pd.concat([preprocessed_df, location_ohe], axis=1)
# preprocessed_df = pd.concat([preprocessed_df, car_type_ohe], axis=1)
# preprocessed_df['car_manufacturer'] = df['car_manufacturer']
preprocessed_df
```

```python
def heatmap(data: pd.DataFrame):
    plt.figure(figsize=(20, 20))
    sns.heatmap(data.corr(method='kendall'), square=True, linecolor='white', annot=True)
    plt.xticks(rotation=45)
    plt.show()
```

```python
def sort_df_by_corr(data: pd.DataFrame, column: str):
    return data.corr(method='kendall')[column].sort_values(ascending=False)
```

```python
def delete_columns(data: pd.DataFrame, columns: list):
    return data.drop(columns, axis=1)
```
